```python
df3 = ['dbn', 'school_name', 'Enrollment']
values = [0, 55,55]

plt.figure(figsize=(9,3))

plt.subplot(131)
plt.bar(df3, values)
plt.subplot(132)
plt.scatter(df3, values)
plt.subplot(133)
plt.plot(df3, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
numberOfPoints = 11
rangeX = np.linspace(0, 50, numberOfPoints)
df3 = np.random.rand(numberOfPoints-1) * 4.0
barWidth = 3.0
middleValue = (rangeX[0:-1] +
            rangeX[1:]) * 0.5 - barWidth * 0.5
plt.bar(middleValue,
        df3,
        facecolor='chocolate',
        width=barWidth,
        label='barChart')
plt.xticks(middleValue)
plt.xlabel('aca_tot_09')
plt.ylabel('Students received surveys')
plt.title('School Type')
plt.grid(True)
plt.legend()
plt.show()
```
